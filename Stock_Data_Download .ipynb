```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import sys
from dateutil.relativedelta import relativedelta
import pandas_datareader.data as web
```

```python
#an excel of various Yahoo Finance Tickers
e=pd.read_excel('Yahoo_Ticker_2017.xlsx')
```

```python
#Skip irrelevent columns and rows in the .xlsx file
e.columns=e.loc[2]
c=e.columns[:5].values
tickers=e[c].loc[3:]
```

```python
###locate US national stocks and reindex
USStocks=tickers.loc[tickers['Country'] == 'USA']
USStocks=USStocks.loc[USStocks['Exchange']=='NMS']
USStocks.index=np.linspace(0,1782,1783,dtype='int')
```

```python
USStocks
```

```python
#get any stock whose data is available from 2013-12-1 to 2019-12-30
d={}
start_time=datetime.datetime(2019,12,1)- relativedelta(years=6)
end_time=datetime.datetime(2019,12,30)
for ticker in USStocks['Ticker']:
  try:     
    a=web.get_data_yahoo(ticker,start=start_time,end=end_time)
  
    if len(a.index)==1530:   
      d[ticker]=a
      
    else:
      print(ticker,'incomplete data')
  except:
    print(ticker,'failed')
       
```

```python
psave(d,'StockPool')
```
